```python
def write_files(filename: str, keywords: list):
    # with open(f"product_categories/{filename}.csv", "w") as f:
    with open(f"product_categories_parent/{filename}.csv", "w") as f:
        f.write("keywords\n")
        for i in keywords[:-1]:
            f.write(
                i.replace(',', ' ')+'\n'
            )
        f.write(
                keywords[-1]
            )
```
