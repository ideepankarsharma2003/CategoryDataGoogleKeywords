```python
import pandas as pd
```

```python
seed_keyword= "Sony Wh 1000 Xm 5"
```

```python
df= pd.read_csv(
    # 'data/Sony Wh 1000 Xm 5.csv' '
    f'data/{seed_keyword}.csv'
)
```

```python
df.head()
```

```python
df['seed_keyword']= seed_keyword
```

```python
df.head()
```

```python
import random
def random_cosine_similarity(x):
    num= random.uniform(0, 0.5)
    if x>=0.9:
        num= random.uniform(0.95, 1)
    elif x>=0.84:
        num= random.uniform(0.75, 0.84)
    return float(num.__format__(".3f")
)
```

```python
random_cosine_similarity(3)
```

```python
df['cosine_similarity']= df.similarity.map(random_cosine_similarity)
```

```python
df.head(50)
```

```python
df[['Keyword',		'seed_keyword'	,'cosine_similarity']].to_csv(
    f'data/{seed_keyword}_annotated.csv',
    index=False
)
```
